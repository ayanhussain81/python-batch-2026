Check For Missing Values

```python
print("Missing Values per column")
print(df.isnull().sum())
```

```python
# value /   * 100
print("Missing % per column")
print((df.isnull().sum() / len(df) * 100).round(1))
```

```python
# median -- Mid values 
# 1,2,3,4,5,6,7

median_age = df["Age"].median()
print("Median Age! ", median_age)
```

```python
# lets fill the null values with median age 

df["Age"] = df["Age"].fillna(median_age)
print(df.isnull().sum())
```

```python
df.describe()
```

```python
# Fix Cabin by dropping column
df = df.drop(columns=["Cabin"])
print(df.columns.to_list())
```

```python
df.shape
```

```python
# mean median mode 

# To fill the embark, we will use mode 

most_common = df['Embarked'].mode()[0]
most_common
```

```python
df['Embarked'].value_counts()
```

```python
df["Embarked"] = df['Embarked'].fillna(most_common)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=['PassengerId','Name','Ticket'])
print(df.shape)
```

```python
df.to_csv('titanic_clean.csv', index=False)
```

```python
# lets check does fare matter 

firs_class = df[df['Pclass'] == 1]
third_class = df[df['Pclass'] == 3]

first_rate = firs_class["Survived"].mean() * 100
third_rate = third_class["Survived"].mean() * 100

print(f"First Class Rate: {first_rate}")
print(f"Third Class Rate: {third_rate}")
```

```python
# lets Gender Ratio in Survival Rate

female = df[df['Sex']=='female']
male = df[df['Sex'] == 'male']

print(f'Female Survival Rate: {female['Survived'].mean()*100}')
print(f'Male Survival Rate: {male['Survived'].mean()*100}')
```

```python
# youngest and olded survivors 
# Most expensive tickets
# childrens (age <10)
# Females who survived belongs from first class
```

Group By Analysis

```python
# Survival rate by class:

class_survival = df.groupby('Pclass')['Survived'].mean() * 100
print(class_survival.round(2))
```

```python
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print(gender_survival.round(2))
```

```python
# gender survival ratio by class 

combined = df.groupby(["Pclass","Sex"])['Survived'].mean() * 100
print(combined)
```
