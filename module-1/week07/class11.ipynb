```python
# Plain Python way — painful
with open("students.csv", "r") as f:
    lines = f.readlines()

headers = lines[0].strip().split(",")
students = []
for line in lines[1:]:
    values = line.strip().split(",")
    students.append(dict(zip(headers, values)))

# Just to get average score:
scores = [int(s["score"]) for s in students]
avg = sum(scores) / len(scores)
```

```python
df["Score_double"] = df["Scores"] * 3
print(df)
```

### Sorting 

```python
# Ascending 
df_sorted =  df.sort_values("Scores")
print(df_sorted)
```

```python
# Descending 
df_sorted =  df.sort_values("Scores", ascending=False) 
print(df_sorted)
```

```python
# sort by multiple columns 
df_sorted = df.sort_values(["Scores","Age"],ascending = [False,False])
print(df_sorted)
```
